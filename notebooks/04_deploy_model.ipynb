```python
import nest_asyncio #allows nested event loops in Jupyter.
import uvicorn
from taxi_ride.models import deploy_model as deploy

# Fix event loop in Jupyter
nest_asyncio.apply()

# Configure server
config = uvicorn.Config(deploy.app, host="0.0.0.0", port=9696, log_level="info")
server = uvicorn.Server(config)

# Start the server in the notebook (non-blocking)
import threading

thread = threading.Thread(target=server.run)
thread.start()
```
